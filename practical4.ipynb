```python
import numpy as np

# Simulate a coin toss experiment
num_tosses = 1000  # Number of coin tosses
coin_outcomes = ['Heads', 'Tails']
tosses = np.random.choice(coin_outcomes, num_tosses)

# Count occurrences of Heads and Tails
heads_count = np.sum(tosses == 'Heads')
tails_count = np.sum(tosses == 'Tails')

# Compute probabilities
probability_heads = heads_count / num_tosses
probability_tails = tails_count / num_tosses

print(f"Total tosses: {num_tosses}")
print(f"Heads count: {heads_count}")
print(f"Tails count: {tails_count}")
print(f"Probability of Heads: {probability_heads:.4f}")
print(f"Probability of Tails: {probability_tails:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the normal distribution
mu, sigma = 0, 0.1 # mean and standard deviation
s = np.random.normal(mu, sigma, 1000) # 1000 samples

# Create the histogram
count, bins, ignored = plt.hist(s, 30, density=True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - mu)**2 / (2 * sigma**2) ), linewidth=2, color='r')
plt.xlabel('Value')
plt.ylabel('Probability')
plt.title('Histogram of a Normal Distribution')
plt.grid(True)
plt.show()
```
