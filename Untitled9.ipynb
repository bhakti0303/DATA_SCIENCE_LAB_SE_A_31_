### Handling Missing Values

Based on the `df.info()` and `df.isnull().sum()` outputs, several columns have 20 missing values. Since this is a small percentage of the total dataset (20 out of 506 rows), I will drop these rows to ensure data quality for the regression model.

```python
# Drop rows with any missing values
df_cleaned = df.dropna()

print(f"Original number of rows: {df.shape[0]}")
print(f"Number of rows after dropping missing values: {df_cleaned.shape[0]}")

display(df_cleaned.head())
```

### Handling Duplicate Rows

Next, I will check for and remove any duplicate rows in the cleaned dataset to ensure each observation is unique.

```python
# Check for duplicate rows
duplicates_count = df_cleaned.duplicated().sum()
print(f"Number of duplicate rows found: {duplicates_count}")

# If duplicates exist, remove them
if duplicates_count > 0:
    df_cleaned = df_cleaned.drop_duplicates()
    print(f"Number of rows after removing duplicates: {df_cleaned.shape[0]}")
else:
    print("No duplicate rows found.")

display(df_cleaned.head())
```
