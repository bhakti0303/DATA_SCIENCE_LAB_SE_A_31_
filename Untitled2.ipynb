## Minimax Algorithm

The Minimax algorithm is a decision-making algorithm used in artificial intelligence, game theory, and robotics. It is particularly used for two-player zero-sum games (where one player's gain is another's loss) such as chess, checkers, and Tic-Tac-Toe. The goal of Minimax is to find the optimal move for a player, assuming that the opponent also plays optimally.

The algorithm works by recursively exploring the game tree, evaluating each possible move and its consequences. It assigns a score to each terminal (game-over) state. For non-terminal states, it alternates between two phases:

*   **Maximizer's Turn (Max Player):** The player tries to maximize the score, choosing the move that leads to the highest possible score from its children.
*   **Minimizer's Turn (Min Player):** The player tries to minimize the score, choosing the move that leads to the lowest possible score from its children.

The recursion continues until a predefined depth limit is reached or a terminal state is found. The result is the best possible score that the maximizer can achieve against an optimal minimizer, and vice-versa.

```python
def minimax(node, depth, maximizing_player, game_tree):
    # Base case: Leaf node or depth limit reached
    if depth == 0 or node not in game_tree:
        # If it's a leaf node, its value is defined directly
        # Otherwise, for simplicity, we assume some heuristic evaluation here
        # For this example, leaf nodes are the only ones with direct values
        return game_tree.get(node, node) # node itself is the score if not in tree (i.e., a leaf)

    if maximizing_player:
        max_eval = float('-inf')
        # Iterate over children of the current node
        for child in game_tree.get(node, []):
            # Recursively call minimax for the child, switching player
            eval = minimax(child, depth - 1, False, game_tree)
            max_eval = max(max_eval, eval)
        return max_eval
    else:
        min_eval = float('inf')
        # Iterate over children of the current node
        for child in game_tree.get(node, []):
            # Recursively call minimax for the child, switching player
            eval = minimax(child, depth - 1, True, game_tree)
            min_eval = min(min_eval, eval)
        return min_eval

# Example Game Tree:
# This represents a simple game where 'A' is the root, and numbers are terminal scores.
# Max player wants to maximize the score, Min player wants to minimize it.
# Depth 0 nodes are the terminal states with scores.
# Format: {parent_node: [child1, child2, ...]},
# where child can be another node or a terminal score (int/float)
game_tree = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F', 'G'],
    'D': [3, 5], # Terminal nodes
    'E': [2, 9], # Terminal nodes
    'F': [1, 0], # Terminal nodes
    'G': [7, 4]  # Terminal nodes
}

# Let's assume the Max player starts and the total depth of the game tree is 3 from root 'A'
# The game ends after 3 moves, reaching the terminal states.
starting_node = 'A'
game_depth = 3 # Max depth from root to a leaf

# The minimax function for the starting player (Maximizing Player)
optimal_value = minimax(starting_node, game_depth, True, game_tree)

print(f"The optimal value for the Max player from node '{starting_node}' is: {optimal_value}")

# Another example with a slightly different tree structure
game_tree_2 = {
    'Root': ['L1A', 'L1B'],
    'L1A': ['L2A', 'L2B'],
    'L1B': ['L2C', 'L2D'],
    'L2A': [3, 12],
    'L2B': [8, 2],
    'L2C': [14, 5],
    'L2D': [2, 4]
}

optimal_value_2 = minimax('Root', 3, True, game_tree_2)
print(f"The optimal value for the Max player from node 'Root' in game_tree_2 is: {optimal_value_2}")
```

## A* Search Algorithm

A* is a popular pathfinding algorithm used in many areas of computer science, including artificial intelligence and robotics. It is a 'best-first' search algorithm that finds the shortest path between a starting node and a goal node in a graph. It does this by using a heuristic function to estimate the cost from the current node to the goal, and combining it with the actual cost from the start node to the current node.

The algorithm uses two primary data structures:
*   **Open List (Priority Queue):** Stores nodes that have been visited but whose neighbors have not yet been fully explored. Nodes are ordered by their f-cost (g-cost + h-cost).
*   **Closed List (Set):** Stores nodes that have already been fully evaluated.

The cost functions are:
*   **g-cost:** The actual cost from the start node to the current node.
*   **h-cost:** The estimated cost (heuristic) from the current node to the goal node.
*   **f-cost:** The total estimated cost of the path through the current node (g-cost + h-cost).

```python
import heapq

def a_star(graph, start, goal, heuristic):
    # The open list (priority queue) stores (f_cost, node, path)
    open_list = [(0 + heuristic(start, goal), start, [start])]
    heapq.heapify(open_list)

    # g_costs stores the actual cost from start to a node
    g_costs = {start: 0}

    # closed_list stores nodes already evaluated
    closed_list = set()

    while open_list:
        # Pop the node with the lowest f_cost
        f_cost, current_node, path = heapq.heappop(open_list)

        if current_node in closed_list:
            continue

        closed_list.add(current_node)

        if current_node == goal:
            return path, g_costs[goal] # Return the path and its total cost

        for neighbor, weight in graph.get(current_node, []):
            # Calculate tentative g_cost for the neighbor
            tentative_g_cost = g_costs[current_node] + weight

            if neighbor not in g_costs or tentative_g_cost < g_costs[neighbor]:
                g_costs[neighbor] = tentative_g_cost
                f_cost_neighbor = tentative_g_cost + heuristic(neighbor, goal)
                heapq.heappush(open_list, (f_cost_neighbor, neighbor, path + [neighbor]))

    return None, None # No path found

# Example Graph (Adjacency List with weights)
# Format: {node: [(neighbor, weight), ...]},
# Weights represent the cost to move from node to neighbor
graph_weighted = {
    'A': [('B', 1), ('C', 4)],
    'B': [('A', 1), ('D', 2), ('E', 5)],
    'C': [('A', 4), ('F', 1)],
    'D': [('B', 2)],
    'E': [('B', 5), ('G', 2)],
    'F': [('C', 1), ('G', 3)],
    'G': [('E', 2), ('F', 3)]
}

# Heuristic function (e.g., Euclidean distance or Manhattan distance for grid)
# For a generic graph, we can use a simple (often less informative) heuristic
# For demonstration, let's assume some pre-calculated 'straight-line' distances to the goal.
# In a real scenario, this would be based on spatial coordinates.
heuristic_values = {
    'A': {'G': 7},
    'B': {'G': 6},
    'C': {'G': 4},
    'D': {'G': 8},
    'E': {'G': 3},
    'F': {'G': 2},
    'G': {'G': 0}
}

def heuristic(node, goal):
    return heuristic_values.get(node, {}).get(goal, float('inf'))

start_node = 'A'
goal_node = 'G'

path, cost = a_star(graph_weighted, start_node, goal_node, heuristic)

if path:
    print(f"A* Path from {start_node} to {goal_node}: {path}")
    print(f"Total cost: {cost}")
else:
    print(f"No path found from {start_node} to {goal_node}.")
```

## Breadth-First Search (BFS)

BFS is an algorithm for traversing or searching tree or graph data structures. It starts at the tree root (or some arbitrary node of a graph, sometimes referred to as a 'search key') and explores all of the neighbor nodes at the present depth prior to moving on to the nodes at the next depth level.

It uses a queue data structure to store the nodes to be visited.

```python
from collections import deque

def bfs(graph, start_node):
    visited = set()  # To keep track of visited nodes
    queue = deque([start_node])  # Initialize queue with the start node
    visited.add(start_node)

    bfs_traversal = []

    while queue:
        current_node = queue.popleft()  # Dequeue a node
        bfs_traversal.append(current_node)

        for neighbor in graph.get(current_node, []):
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(neighbor)  # Enqueue unvisited neighbors
    return bfs_traversal

# Example Graph Representation (Adjacency List)
graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F', 'G'],
    'D': ['B'],
    'E': ['B', 'H'],
    'F': ['C'],
    'G': ['C'],
    'H': ['E']
}

print("BFS Traversal (starting from A):")
print(bfs(graph, 'A'))
```

## Depth-First Search (DFS)

DFS is an algorithm for traversing or searching tree or graph data structures. The algorithm starts at the root node (selecting some arbitrary node as the root node in the case of a graph) and explores as far as possible along each branch before backtracking.

It typically uses recursion or an explicit stack data structure.

```python
def dfs(graph, start_node, visited=None):
    if visited is None:
        visited = set()  # To keep track of visited nodes

    dfs_traversal = []
    stack = [start_node] # Initialize stack with the start node

    while stack:
        current_node = stack.pop()
        if current_node not in visited:
            visited.add(current_node)
            dfs_traversal.append(current_node)

            # Push unvisited neighbors onto the stack
            # It's important to push in reverse order to explore in a consistent manner
            # or simply rely on the graph structure if order doesn't matter
            for neighbor in reversed(graph.get(current_node, [])):
                if neighbor not in visited:
                    stack.append(neighbor)
    return dfs_traversal

# Recursive DFS implementation (often more intuitive)
def dfs_recursive(graph, start_node, visited=None, traversal_order=None):
    if visited is None:
        visited = set()
    if traversal_order is None:
        traversal_order = []

    visited.add(start_node)
    traversal_order.append(start_node)

    for neighbor in graph.get(start_node, []):
        if neighbor not in visited:
            dfs_recursive(graph, neighbor, visited, traversal_order)
    return traversal_order

# Using the same graph as BFS example
print("DFS Traversal (iterative, starting from A):")
print(dfs(graph, 'A'))

print("DFS Traversal (recursive, starting from A):")
print(dfs_recursive(graph, 'A'))
```
